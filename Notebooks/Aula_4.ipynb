### Exemplo: arquivo csv

**Atenção:** Os exemplos abaixo foram desenvolvidos em um ambiente Linux; caso você utilize Windows ou MacOS, deve alterar o caminho do arquivo de acordo com a convenção do seu sistema operacional.

Para lermos arquivos .csv (Comma Separated Values, ou "valores separados por vírgula") podemos utilizar o módulo csv, como no exemplo abaixo, em que o arquivo é lido de uma só vez através do comando csv.reader(arquivo), e em seguida imprimimos cada linha do arquivo separadamente. Perceba que o objeto reader é iterável, ou seja, está em formato próprio para que possamos ler cada linha do arquivo em um for sem utilizarmos índices.

```python
import csv
with open('../exemplos/exemplo_3/MCT II Rio Chui Final.csv', 'r') as arquivo:
    reader = csv.reader(arquivo)
    for row in reader:
        print(", ".join(row))
```

Observe também que cada linha, neste caso, é uma lista de strings.

Agora, suponha que desejamos retornar um dicionário para cada linha, ao invés de uma lista de strings. Isso pode ser útil caso tenhamos um arquivo com um cabeçalho descrevendo os nomes dos campos, como é o caso do arquivo com o qual estamos trabalhando.

Neste caso, usamos o método csv.DictReader :

```python
import csv
with open('../exemplos/exemplo_3/MCT II Rio Chui Final.csv', 'r') as arquivo:
    reader = csv.DictReader(arquivo)
    for row in reader:
        print(row)
```

Uma das vantagens desta formulação é que podemos acessar as colunas de dados de maneira bastante eficiente. Por exemplo, suponha que queremos imprimir apenas os valores referentes às colunas "Estação" e "Prof. (m)" da nossa tabela.

```python
import csv
with open('../exemplos/exemplo_3/MCT II Rio Chui Final.csv', 'r') as arquivo:
    reader = csv.DictReader(arquivo)
    for row in reader:
        print(row['Estação'], row['Prof. (m)'])
```

# Pandas

A biblioteca mais apropriada para lidarmos com grandes arquivos de texto, estilo csv ou txt, é a Pandas
http://pandas.pydata.org

Observe que o arquivo abaixo tem uma visualização confusa justamente por seu tamanho:

```python
with open('../exemplos/exemplo_4/table_w_tax.txt', 'r') as arquivo:
    for linha in arquivo:
        print(linha)
```

Por outro lado, ao importarmos a biblioteca pandas, podemos usar o método read_csv. Imediatamente parece que o arquivo se transforma em uma tabela de fácil visualização:

```python
import pandas as pd
dados = pd.read_csv('../exemplos/exemplo_4/table_w_tax.txt')
```

```python
dados
```

Na verdade, dados é um tipo diferente de estrutura:

```python
type(dados)
```

Um DataFrame é uma estrutura criada pela biblioteca pandas que pode ser considerada como uma tabela, organizada com nomes de colunas e índices de linhas.

No entanto, para o arquivo anterior, ainda temos problemas de leitura. De fato, o arquivo tem colunas separadas pelo caracter TAB. Assim, podemos modificar o comando para obtermos

```python
dados = pd.read_csv('../exemplos/exemplo_4/table_w_tax.txt', delimiter='\t')
```

```python
dados
```

Por fim, para eliminarmos a primeira linha do arquivo já que está não faz parte do cabeçalho em que estamos interessados, modificamos mais uma vez o comando:

```python
dados = pd.read_csv('../exemplos/exemplo_4/table_w_tax.txt', delimiter='\t', header=1)
dados
```

Podemos acessar uma coluna do DataFrame acima (por exemplo, a coluna cujo rótulo é "#OTU ID") usando a seguinte sintaxe:
